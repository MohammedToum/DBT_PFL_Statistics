```python
%%ducksql

/*
"""
~ ############### OTHER FILE  ###############~
#------------------ ## OBJECTIVES ## ------------------#
 - To Do:
    - find unique values for '
        - 'club_involved_name',
        
    - find way to parse
        - 'club_involved_name',

"""
*/

-- 1) INSERT the cleaned rows into your target table
INSERT INTO clubs (raw_club, cleaned_full_name, parent_club, abbreviation, country)
SELECT
    c.raw_club,
    c.cleaned_full_name,
    c.parent_club,
    c.abbreviation,
    c.country
FROM (

  WITH src AS (
    -- replace this table/column with your real source
    SELECT club_involved_name AS raw_club
    FROM staging.other_file_transfers_all
    WHERE club_involved_name IS NOT NULL
  ),

  -- Map common aliases/nicknames/short forms to official names (extend this!)
  club_alias_map AS (
    SELECT * FROM (VALUES
      ('MAN CITY',              'MANCHESTER CITY'),
      ('MAN UTD',               'MANCHESTER UNITED'),
      ('BOR. DORTMUND',         'BORUSSIA DORTMUND'),
      ('BOR. M''GLADBACH',      'BORUSSIA MONCHENGLADBACH'),
      ('BAYERN MUNICH',         'FC BAYERN MUNCHEN'),
      ('RM CASTILLA',           'REAL MADRID CASTILLA'),
      ('BARÇA U19',             'FC BARCELONA U19'),
      ('BARCELONA',             'FC BARCELONA'),
      ('SPURS',                 'TOTTENHAM HOTSPUR'),
      ('QPR',                   'QUEENS PARK RANGERS'),
      ('PSV EINDHOVEN',         'PSV'),
      ('AJAX',                  'AFC AJAX')
    ) AS t(alias_raw, official_name)
  ),

  -- Country mapping (seed a few; grow this over time)
  club_country_map AS (
    SELECT * FROM (VALUES
      ('MANCHESTER CITY',       'England'),
      ('MANCHESTER UNITED',     'England'),
      ('TOTTENHAM HOTSPUR',     'England'),
      ('QUEENS PARK RANGERS',   'England'),
      ('AFC AJAX',              'Netherlands'),
      ('PSV',                   'Netherlands'),
      ('FC BARCELONA',          'Spain'),
      ('REAL MADRID',           'Spain'),
      ('REAL MADRID CASTILLA',  'Spain'),
      ('BORUSSIA DORTMUND',     'Germany'),
      ('BORUSSIA MONCHENGLADBACH','Germany'),
      ('FC BAYERN MUNCHEN',     'Germany')
    ) AS t(cleaned_full_name, country)
  ),

  -- 2) Normalize raw: UPPER+TRIM, strip punctuation variants for matching
  norm AS (
    SELECT
      raw_club,
      UPPER(TRIM(raw_club))                           AS up_raw,
      -- unify a few common punctuation bits for alias matching
      UPPER(TRIM(REGEXP_REPLACE(raw_club, '[.’''’]', '')))
        AS up_nopunct
    FROM src
  ),

  -- 3) Compute the senior/base club name by removing youth/reserve markers
  base_name AS (
    SELECT
      raw_club,
      -- remove youth/reserve patterns
      TRIM(
        REGEXP_REPLACE(
          REGEXP_REPLACE(
            REGEXP_REPLACE(
              REGEXP_REPLACE(
                REGEXP_REPLACE(
                  up_raw,
                  -- drop Uxx squads
                  '\bU(16|17|18|19|20|21|22|23)\b', ''
                ),
                -- drop reserve/second team markers
                '\b(RES(\.|ERVES)?|B|II|A(?:CADEMY|CA)|YOUTH)\b', ''
              ),
              -- collapse duplicate spaces
              '\s+', ' '
            ),
            -- tidy leftover trailing/leading dangles
            '(^\s+|\s+$)', ''
          ),
          -- also tidy some " CF B", " FC U19" style seams
          '\s+(CF|FC|SC)\s*$', '\1'
        )
      ) AS base_up
    FROM norm
  ),

  -- 4) Resolve aliases on the base name (so “MAN UTD U21” collapses to “MANCHESTER UNITED”)
  resolved AS (
    SELECT
      n.raw_club,
      COALESCE(a.official_name, b.base_up) AS cleaned_full_name_up,
      -- is this a youth/reserve row?
      CASE
        WHEN n.up_raw ~ '\bU(16|17|18|19|20|21|22|23)\b' THEN TRUE
        WHEN n.up_raw ~ '\b(RES(\.|ERVES)?|B|II|A(?:CADEMY|CA)|YOUTH)\b' THEN TRUE
        ELSE FALSE
      END AS is_youth_or_res
    FROM norm n
    JOIN base_name b ON b.raw_club = n.raw_club
    LEFT JOIN club_alias_map a
      ON b.base_up = a.alias_raw          -- alias match on normalized base
      OR n.up_nopunct = a.alias_raw       -- fall back if punctuation is the only diff
  ),

  -- 5) Final shape: parent_club = senior name; cleaned_full_name keeps the senior name too
  shaped AS (
    SELECT
      raw_club,
      cleaned_full_name_up                                        AS cleaned_full_name,
      -- parent = senior club for youth/reserve, else self
      CASE WHEN is_youth_or_res THEN cleaned_full_name_up
           ELSE cleaned_full_name_up
      END                                                         AS parent_club,
      -- fallback abbreviation: strip common tokens then take first 3 letters of the squashed name
      CASE
        WHEN LENGTH(REGEXP_REPLACE(cleaned_full_name_up, '\b(FC|CF|SC|AC|AS|CD|UD|US|SD)\b', '', 'g')) >= 3
          THEN SUBSTR(REGEXP_REPLACE(cleaned_full_name_up, '\b(FC|CF|SC|AC|AS|CD|UD|US|SD)\b', '', 'g')
                       |> REGEXP_REPLACE('[^A-Z]', '', 'g')
                       , 1, 3)
        ELSE SUBSTR(REGEXP_REPLACE(cleaned_full_name_up, '[^A-Z]', '', 'g'), 1, 3)
      END                                                         AS abbreviation,
      COALESCE(ccm.country, 'Unknown')                            AS country
    FROM resolved r
    LEFT JOIN club_country_map ccm
      ON r.cleaned_full_name_up = ccm.cleaned_full_name
  )

  SELECT
    raw_club,
    cleaned_full_name,
    parent_club,
    abbreviation,
    country
  FROM shaped

) AS c;
```

```python
%%ducksql
-- Longest-prefix country match via left joins (prints rows reliably)
WITH countries_list(country_name) AS (
  SELECT * FROM (VALUES
    ('Antigua and Barbuda'), ('DR Congo'), ('Costa Rica'), ('Cote d''Ivoire'),
    ('Czech Republic'), ('United States'), ('St. Lucia'), ('St. Kitts & Nevis'),
    ('Guinea-Bissau'), ('Northern Ireland'), ('The Gambia'), ('Cape Verde'),
    ('Isle of Man'), ('North Macedonia')
  )
),
clist AS (
  SELECT
    LOWER(TRIM(REGEXP_REPLACE(country_name, '\s+', ' '))) AS key_name,
    country_name
  FROM countries_list
),
base AS (
  SELECT
    nationality,
    LOWER(TRIM(REGEXP_REPLACE(COALESCE(nationality,''), '\s+', ' '))) AS nat_l
  FROM staging.transfers_all
),
pref AS (
  SELECT
    nationality,
    -- word prefixes (1..4)
    array_to_string(array_slice(str_split(nat_l, ' '), 0, 4), ' ') AS p4,
    array_to_string(array_slice(str_split(nat_l, ' '), 0, 3), ' ') AS p3,
    array_to_string(array_slice(str_split(nat_l, ' '), 0, 2), ' ') AS p2,
    array_to_string(array_slice(str_split(nat_l, ' '), 0, 1), ' ') AS p1
  FROM base
),
sorted AS (
    SELECT
  -- p.nationality,
  LOWER(COALESCE(c4.country_name, c3.country_name, c2.country_name, p.p1)) AS final_country
FROM pref p
LEFT JOIN clist c4 ON p.p4 = c4.key_name
LEFT JOIN clist c3 ON p.p3 = c3.key_name
LEFT JOIN clist c2 ON p.p2 = c2.key_name
ORDER BY p.nationality ASC
)
SELECT 
    DISTINCT final_country 
FROM
    sorted
ORDER BY
    final_country ASC;
```

```python
%%ducksql

/*
"""
#------------------ ## OBJECTIVES ## ------------------#
 - To Do:
    - parse unique values for "nationality"


    # Nations with spaces list:
        Antigua and Barbuda
        DR Congo
        Costa Rica
        Cote d'Ivoire
        Czech Republic
        United States
        St. Lucia
        St. Kitts & Nevis
        Guinea-Bissau
        Northern Ireland
        The Gambia
        Cape Verde
        Isle of Man
        North Macedonia

"""
*/

WITH countries_list(country_name) AS (
  SELECT * FROM (VALUES
    ('Antigua and Barbuda'), ('DR Congo'), ('Costa Rica'), ('Cote d''Ivoire'),
    ('Czech Republic'), ('United States'), ('St. Lucia'), ('St. Kitts & Nevis'),
    ('Guinea-Bissau'), ('Northern Ireland'), ('The Gambia'), ('Cape Verde'),
    ('Isle of Man'), ('North Macedonia')
  )
),
base AS (
  SELECT
    nationality,
    TRIM(REGEXP_REPLACE(
      REGEXP_REPLACE(
        LOWER(REGEXP_REPLACE(COALESCE(nationality,''), '[\u00A0\u2000-\u200F\u202A-\u202E\u2060-\u2069\uFEFF]', '')),
        '\s+', ' '
      ),
      ' +', ' '
    )) AS nat_l
  FROM staging.transfers_all
),
tok AS (
  SELECT
    nationality,
    nat_l,
    str_split(nat_l, ' ') AS words
  FROM base
),
prefixes AS (
  SELECT
    t.nationality,
    array_to_string(array_slice(t.words, 0, n.n), ' ') AS prefix
  FROM tok t
  JOIN range(1, 5) AS n(n)          -- prefixes of 1..4 words
),
clist AS (
  SELECT
    LOWER(TRIM(REGEXP_REPLACE(country_name, '\s+', ' '))) AS key_name,
    country_name
  FROM countries_list
),
matches AS (
  SELECT
    p.nationality,
    c.country_name,
    row_number() OVER (PARTITION BY p.nationality ORDER BY length(c.country_name) DESC) AS rn
  FROM prefixes p
  JOIN clist c ON p.prefix = c.key_name
)
SELECT
  b.nationality,
  coalesce(
    (SELECT m.country_name FROM matches m WHERE m.nationality = b.nationality AND m.rn = 1),
    array_to_string(array_slice(str_split(b.nat_l, ' '), 0, 1), ' ')
  ) AS final_country
FROM base b
LIMIT 50;
```

```python
%%ducksql

/*
"""
#------------------ ## OBJECTIVES ## ------------------#
 - To Do:
    - parse fee value to million decimal point
                ##################
                ###### DONE ######
                ##################       
"""

*/

WITH src AS (
  SELECT
    fee,
    LOWER(COALESCE(fee, '')) AS fee_l
  FROM staging.transfers_all
),
parsed AS (
  SELECT
    fee,
    fee_l,

    /* numeric in MILLIONS */
    CASE
      WHEN fee_l LIKE '%free transfer%' THEN 0.0

      /* '?' or '-' but only when no unit appears */
      WHEN (fee_l LIKE '%?%' OR fee_l LIKE '%-%')
           AND NOT REGEXP_MATCHES(fee_l, '€?\s*[0-9]+(?:[\.,][0-9]+)?\s*(m|k|th\.?)\b')
        THEN NULL

      /* millions: number followed by 'm' (euro/signs/words can be anywhere) */
      WHEN REGEXP_MATCHES(fee_l, '€?\s*[0-9]+(?:[\.,][0-9]+)?\s*m\b') THEN
        CAST(
          REPLACE(
            REGEXP_EXTRACT(fee_l, '€?\s*([0-9]+(?:[\.,][0-9]+)?)\s*m\b', 1),
            ',', '.'
          ) AS DOUBLE
        )

      /* thousands: number followed by 'k' or 'th' -> convert to millions */
      WHEN REGEXP_MATCHES(fee_l, '€?\s*[0-9]+(?:[\.,][0-9]+)?\s*(k|th\.?)\b') THEN
        CAST(
          REPLACE(
            REGEXP_EXTRACT(fee_l, '€?\s*([0-9]+(?:[\.,][0-9]+)?)\s*(?:k|th\.?)\b', 1),
            ',', '.'
          ) AS DOUBLE
        ) / 1000.0

      ELSE NULL
    END AS fee_millions,

    /* reason */
    CASE
      WHEN fee_l LIKE '%free transfer%' THEN 'Free Transfer'
      WHEN (fee_l LIKE '%?%' OR fee_l LIKE '%-%')
           AND NOT REGEXP_MATCHES(fee_l, '€?\s*[0-9]+(?:[\.,][0-9]+)?\s*(m|k|th\.?)\b')
        THEN 'Unknown'
      WHEN REGEXP_MATCHES(fee_l, '€?\s*[0-9]+(?:[\.,][0-9]+)?\s*m\b') THEN 'Millions'
      WHEN REGEXP_MATCHES(fee_l, '€?\s*[0-9]+(?:[\.,][0-9]+)?\s*(k|th\.?)\b') THEN 'Thousands→Millions'
      ELSE 'Dunno'
    END AS reason
  FROM src
)
SELECT fee, fee_millions, reason
FROM parsed
LIMIT 100;
```

```python
%%ducksql
/*
"""
----------------- ## OBJECTIVES ## ------------------
 - To Do:
    - find unique values for:
        -"club_name" <--------
        - "player_name",
        - "nationality",
        - "position"
        -"club_involved_name"
        -"club_involved_country"
"""
*/


WITH norm AS (
  SELECT
    club_name,
    -- remove NBSP, ZWSP/ZWNJ/ZWJ, LRM/RLM, BOM, isolates
    REGEXP_REPLACE(
      club_name,
      '[\u00A0\u2000-\u200F\u202A-\u202E\u2060-\u2069\uFEFF]',
      ''
    ) AS s0
  FROM staging.transfers_all
),
clean AS (
  SELECT
    club_name,
    TRIM(REGEXP_REPLACE(s0, '\s+', ' ')) AS club_name_trimmed,
    LOWER(TRIM(REGEXP_REPLACE(
      REGEXP_REPLACE(s0, '[^a-zA-Z0-9 ]+', ' '),
      '\s+', ' '
    ))) AS club_key
  FROM norm
)
SELECT DISTINCT club_name_trimmed
FROM clean
ORDER BY club_name_trimmed;
```

```python
from pathlib import Path
import duckdb
import pandas as pd

""
----------------- ## OBJECTIVES ## ------------------
 - To Do:
    - find unique values for:
        -"club_name" <--------
            - writes these clubs to a file

"""

# 1) Where to write the seed
SEED_DIR = Path("./dbt/seeds")
SEED_DIR.mkdir(parents=True, exist_ok=True)
OUT = (SEED_DIR / "fpl_2425_clubs.csv").as_posix()

# 2) Build canonical club table for 24/25
con.execute(r"""
CREATE OR REPLACE TEMP VIEW final_fpl_2425_clubs AS
WITH base AS (
  SELECT club_name
  FROM staging.transfers_all
  WHERE LOWER(COALESCE(season,'')) IN ('2023/2024','23/24')
),
norm AS (
  SELECT
    club_name,
    REGEXP_REPLACE(
      club_name,
      '[\u00A0\u2000-\u200F\u202A-\u202E\u2060-\u2069\uFEFF]',
      ''
    ) AS s0
  FROM base
),
clean AS (
  SELECT
    TRIM(REGEXP_REPLACE(s0, '\s+', ' ')) AS club_name_canonical,
    TRIM(
      REGEXP_REPLACE(
        REGEXP_REPLACE(LOWER(s0), '[^a-z0-9 ]+', ' '),
        '\s+', ' '
      )
    ) AS club_key
  FROM norm
)
SELECT
  MIN(club_name_canonical) AS club_name,
  club_key
FROM clean
GROUP BY club_key
ORDER BY club_name;
""")

# 3) Fetch and display result to verify
df = con.sql("SELECT * FROM final_fpl_2425_clubs;").df()
print(f"\nPreview ({len(df)} rows):")
display(df.head(30))  # shows first 30 clubs

# 4) Write only if data is present
if df.empty:
    print("⚠️ No rows returned — check your season filter or source table.")
else:
    con.execute(f"""
    COPY (SELECT * FROM final_fpl_2425_clubs)
    TO '{OUT}'
    (HEADER, DELIMITER ',');
    """)
    print(f"✅ Wrote {len(df)} rows to {OUT}")
```
